```python
### your device should  already have python, numpy and open cv library installed
#   or you can run this in google colab 
#   for open cv : pip install opencv -python


import cv2 as cv    # importing opencv library
import numpy as np  # importing numpy library
a=[]                 # creating an empty array
```

```python
capture = cv.VideoCapture(0)     # this will start your web cam

while True:                                    # runing while loop for capturing video 
    isTrue , frame = capture.read()             # frame will have frames from video
    a.append(frame)                             # appending frame (.png) data (RGB values) in array
    cv.imshow('video',frame)                    # this will open new window showing your video
    
    #important for stoping video
    # it will stop video after 20 sec or if you press s it will stop
    if cv.waitKey(20) & 0xFF == ord('s'):
        break

capture.release()                       # stop web cam
cv.destroyAllWindows()                  # close all video opened using opencv 
```

```python
imgs =np.array(a)                       # importing a into numpy array for easy accessing  
```
